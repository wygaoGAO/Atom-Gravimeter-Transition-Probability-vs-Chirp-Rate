```python
import numpy as np
import matplotlib.pyplot as plt

# 定义参数
g = 9.81  # 局部重力加速度 (m/s^2)
T = 0.3  # 脉冲分离时间 (s)
k_eff = 2 * np.pi / 780e-9  # 有效波矢 (1/m)
C = 0.9  # 可见度
alpha = np.linspace(-150, 150, 1000)  # 频率调节速率 (Hz)

# 计算相位偏移 Δφ
delta_phi = (k_eff * g - alpha) * T**2

# 计算概率 P
P = (1 + C * np.cos(delta_phi)) / 2

# 绘制结果
plt.figure(figsize=(10, 6))
plt.plot(alpha, P)
plt.title('Interference Fringe')
plt.xlabel('Chirp rate (Hz)')
plt.ylabel('Transition Probability P')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 定义参数
g = 9.81  # 局部重力加速度 (m/s^2)
T = 0.3  # 脉冲分离时间 (s)
k_eff = 2 * np.pi / 780e-9  # 有效波矢 (1/m)
C = 0.9  # 可见度
alpha = np.linspace(-150, 150, 1000)  # 频率调节速率 (Hz)

# 定义振动噪声参数
vibration_noise_amplitude = 0.1  # 振动噪声幅度
vibration_noise = np.random.normal(0, vibration_noise_amplitude, alpha.shape)  # 振动噪声

# 计算相位偏移 Δφ
delta_phi = (k_eff * g - alpha) * T**2 + vibration_noise

# 计算概率 P
P = (1 + C * np.cos(delta_phi)) / 2

# 绘制结果
plt.figure(figsize=(10, 6))
plt.plot(alpha, P)
plt.title('Interference Fringe with Vibration Noise')
plt.xlabel('Chirp rate (Hz)')
plt.ylabel('Transition Probability P')
plt.grid(True)
plt.show()
```
