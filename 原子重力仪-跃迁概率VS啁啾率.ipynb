```python
import numpy as np
import matplotlib.pyplot as plt

# 
g = 9.81  #  (m/s^2)
T = 0.3  #  (s)
k_eff = 2 * np.pi / 780e-9  #  (1/m)
C = 0.9  # 
alpha = np.linspace(-150, 150, 1000)  #  (Hz)

#  
delta_phi = (k_eff * g - alpha) * T**2

#  P
P = (1 + C * np.cos(delta_phi)) / 2

# 
plt.figure(figsize=(10, 6))
plt.plot(alpha, P)
plt.title('Interference Fringe')
plt.xlabel('Chirp rate (Hz)')
plt.ylabel('Transition Probability P')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 
g = 9.81  #  (m/s^2)
T = 0.3  #  (s)
k_eff = 2 * np.pi / 780e-9  #  (1/m)
C = 0.9  # 
alpha = np.linspace(-150, 150, 1000)  #  (Hz)

# 
vibration_noise_amplitude = 0.1  # 
vibration_noise = np.random.normal(0, vibration_noise_amplitude, alpha.shape)  # 

#  
delta_phi = (k_eff * g - alpha) * T**2 + vibration_noise

#  P
P = (1 + C * np.cos(delta_phi)) / 2

# 
plt.figure(figsize=(10, 6))
plt.plot(alpha, P)
plt.title('Interference Fringe with Vibration Noise')
plt.xlabel('Chirp rate (Hz)')
plt.ylabel('Transition Probability P')
plt.grid(True)
plt.show()
```
